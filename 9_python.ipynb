```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data from the CSV file Clean Alternative Fuel Vehicle (CAFV) Eligibility
df = pd.read_csv('https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD')
```

```python
df.head(2)
```

```python
# Get all the cars and their types that do not qualify for clean alternative fuel vehicle
    non_clean_cars = df.loc[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] != 'Clean Alternative Fuel Vehicle Eligible', ['Make', 'Model']]
```

```python
# Get all TESLA cars with the model year, and model type made in Bothell City
tesla_cars_bothell = df.loc[(df['Make'] == 'TESLA') & (df['City'] == 'Bothell'), ['Model Year', 'Model']]
```

```python
# Get all the cars that have an electric range of more than 100 and were made after 2015
electric_cars = df.loc[(df['Electric Range'] > 100) & (df['Model Year'] > 2015), :]
```

```python
# Perform data analysis and answer the questions


# Print the results
print("Cars and their types that do not qualify for clean alternative fuel vehicle:")
print(non_clean_cars)

print("\nTESLA cars with the model year and model type made in Bothell City:")
print(tesla_cars_bothell)

print("\nCars that have an electric range of more than 100 and were made after 2015:")
print(electric_cars)
```
